```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
img = cv2.imread('rmstu1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
noisy = np.zeros(img.shape, np.uint8)

p = 0.2

# transvering image
for i in range(img.shape[0]):  
    for j in range(img.shape[1]):  
        r = random.random()
        if r < p / 2:
            noisy[i][j] = [0, 0, 0]  
        elif r < p:
            noisy[i][j] = [255, 255, 255]  
        else:
            noisy[i][j] = img[i][j] 
```

```python
# median blur
denoised = cv2.medianBlur(noisy, 5)

output = [noisy, denoised]

titles = ['Noisy', 'Denoised']

for i in range(2):
    plt.subplot(1, 2, i + 1)
    plt.imshow(output[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
